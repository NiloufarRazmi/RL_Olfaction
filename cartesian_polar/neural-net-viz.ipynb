```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def draw_layer(ax, n_nodes, x, node_radius=0.02, highlight_indices=None, highlight_color='red'):
    """Draws a vertical layer of neurons at position x."""
    y_spacing = 1.0 / (n_nodes + 1)
    for i in range(n_nodes):
        y = 1 - (i + 1) * y_spacing
        color = highlight_color if highlight_indices and i in highlight_indices else 'lightgray'
        circle = plt.Circle((x, y), node_radius, color=color, ec='black')
        ax.add_patch(circle)
```

```python
def draw_network(n_input, n_hidden, highlight_input=[], highlight_hidden=[]):
    fig, ax = plt.subplots(figsize=(10, 30))
    ax.set_aspect('equal')
    ax.axis('off')

    # Draw input and hidden layers
    draw_layer(ax, n_input, x=0.2, highlight_indices=highlight_input)
    draw_layer(ax, n_hidden, x=0.8, highlight_indices=highlight_hidden)

    plt.title("Neural Network Diagram")
    plt.show()
```

```python
draw_network(n_input=19, n_hidden=50, highlight_input=[0], highlight_hidden=[51])
```

```python
# Create a directed graph
G = nx.DiGraph()

# Define layers
layers = [19, 50, 50, 50, 50, 3]  # example layer sizes
# we'll call every 10 nodes 100
positions = {}  # for storing node positions
layer_spacing = 1
node_spacing = 1

node_id = 0
node_labels = {}

for i, num_nodes in enumerate(layers):
    for j in range(num_nodes):
        G.add_node(node_id)
        # Vertically align each layer, center it around y=0
        x = i+5
        y = -j * node_spacing + (num_nodes - 1) * node_spacing / 2
        positions[node_id] = (x, y)
        node_labels[node_id] = str(node_id)
        node_id += 1

# Draw nodes with color control
node_colors = []
for node in G.nodes:
    if node in [0, 23, 32, 36, 86, 88, 90, 134, 139, 206, 220]:
        node_colors.append("green")  # Highlight node 0
    elif node in [219]:
        node_colors.append('red')
    elif node in [221]:
        node_colors.append('green')
    else:
        node_colors.append("lightgray")

plt.figure(figsize=(12, 12))

G.add_edge(0, 23) # 56
G.add_edge(0, 32) # 147
G.add_edge(0, 36) # 180

G.add_edge(23, 86) # 180
G.add_edge(23, 88) # 203
G.add_edge(23, 90) # 220
#G.add_edge(50, 80, width=1)

G.add_edge(88, 134) # 166
G.add_edge(90, 139) # 217

G.add_edge(139, 206) # 381

G.add_edge(206, 220) # turn left

G.add_edge(206, 219)

edge_colors = []
for edge in G.edges():
    if edge in [(0,23), (0,32), (0,36), (23,86), (23,88), (23,90), (88,134), (90,139), (139,206), (206,220)]:
        edge_colors.append('green')
    else:
        edge_colors.append('red')



nx.draw_networkx_nodes(G, positions, node_color=node_colors, node_size=100)
nx.draw_networkx_edges(G, positions, alpha=0.3, edge_color=edge_colors, width=5)
#nx.draw_networkx_labels(G, positions, labels=node_labels, font_size=6)

plt.axis("off")
plt.title("Odor Input: Path of Largest Positive Weight")
plt.savefig('figures/20runs1EWodorHweightpath.png', dpi=500)
plt.show()
```
